```python
# Importing modules
import pandas as pd
import os
import plotly.express as px

data = pd.read_csv('movies_filled_cleaned_processed.csv')
# Print head
data.head()
```

```python
data.groupby(by = ["Distributor"])['World Sales (in $)'].sum().reset_index()
px.pie(data
       ,values = 'World Sales (in $)'
       ,names = "Distributor"
       ,title = 'Percentage of Different Distributor in Total world Sales'
       ,color_discrete_sequence = px.colors.sequential.RdBu
       ,height = 600)
```

```python
data.groupby(by = ["Distributor"])['Domestic Sales (in $)'].sum().reset_index()
px.pie(data
       ,values = 'Domestic Sales (in $)'
       ,names = "Distributor"
       ,title = 'Percentage of Different Distributor in Domestic Sales (in $)'
       ,color_discrete_sequence = px.colors.sequential.RdBu
       ,height = 600)
```

```python
data.groupby(by = ["Distributor"])['International Sales (in $)'].sum().reset_index()
px.pie(data
       ,values = 'International Sales (in $)'
       ,names = "Distributor"
       ,title = 'Percentage of Different Distributor in International Sales (in $)'
       ,color_discrete_sequence = px.colors.sequential.RdBu
       ,height = 600)
```
